Bei konventionellen, verbrennungsmotorisch angetriebenen Fahrzeugen wird das Thermo management üblicherweise im Zusammenhang mit Wirkungsgradverbesserungen und der Abgas nachbehandlung betrachtet. Durch die Elektrifizierung des Antriebsstrangs kommen neue Herausforderungen hinzu. Zum einen muss die temperaturempfindliche Lithium-IonenBatterie, siehe Kapitel 3.2 dieser Reihe, auf Zieltemperaturen gekühlt werden, was eine Anbindung der Batteriekühlung an die Klimatisierung des Fahrzeugs notwendig macht. Zum anderen stellen die Klimatisierung im Sommer und vor allem die Innenraumheizung im Winter neue Ansprüche an die Auslegung dieser Systeme. Mit steigender Elektrifizierung des Antriebsstrangs muss die hierfür erforderliche Leistung mehr und mehr von der Batterie zur Verfügung gestellt werden, mit entsprechender Reduktion der elektrischen Fahrzeugreichweite. Diese Thermomanagementsysteme müssen also möglichst effizient arbeiten. Die Klimatisierung ist jedoch nicht nur unter dem Komfortaspekt und damit der Kundenakzeptanz zu sehen; es spielen hierbei auch Sicherheitsthemen eine wichtige Rolle. Im Winter müssen die Scheiben möglichst schnell und dauerhaft eis- und beschlagfrei gehalten werden, im Sommer wird durch einen angenehm klimatisierten Fahrzeuginnenraum der sogenannte thermische Stress für den Fahrer verringert und seine Konzentrationsfähigkeit erhöht.

传统的燃油车里的热管理通常是为了改进效率和尾气处理。 而在混动和纯电系统里，热管理有了新的定义：一方面为了将温度敏感的锂电池维持在一定的温度， 电池的冷却系统要能可靠高效地调节温度。另一方面，冬夏的车内空调系统也对热管理提出了更高的要求。在电动车里，用电设备往往只能由电池供电，这必将导致车辆续航里程的减少。因此，热管理系统必须被设计得更为高效，使电动车达到和燃油车一样的舒适性和安全性。

### 电动车中的热流

为了能量化地分析对加热和冷却的要求，以及估计这些要求对电池续航里程的影响，就必须分别对车辆在冬季和夏季行驶中的热流进行分析。 下图简要地描述了一辆纯电驱动的车辆在冬季所需的热流。


Bei diesen Bedingungen werden für den eigentlichen Vortrieb des Fahrzeugs im Stadtverkehr etwa 2,5 kW benötigt. Für die Aufrechterhaltung einer Innenraumtemperatur von 22 °C bei Beheizung mit Frischluft ist jedoch eine nahezu doppelt so hohe Leistung von 4,5 kW erforderlich. Von der Batterie und vom Antriebsstrang werden lediglich 0,4 kW Abwärme bereitgestellt. Die notwendige hohe Heizleistung setzt sich aus den Konvektions- und Strahlungsverlusten über die Kabinenfläche (etwa 2 kW) sowie aus dem Wärmeverlust durch die Innenraumabluft (etwa 2,5 kW) zusammen. Dabei muss sowohl während der Aufheizung nach dem Start (in einem nicht vorkonditionierten Fahrzeug, das heißt Innentemperatur gleich Außentemperatur) als auch im stationären Fall, je nach Außentemperatur, bis zu 40 K Temperaturdifferenz ausgeglichen werden, um im Fahrzeuginnenraum die Komforttemperatur von 22 °C zu erreichen beziehungsweise zu halten. In Abschnitt 4 wird zudem noch erläutert, dass Zelltemperaturen unterhalb von 0 °C die Batterie sowohl in der Leistung als auch in der Lebensdauer beeinträchtigen; sie muss deshalb in den kalten Monaten aktiv beheizt werden. 

纯电车在冬季城市交通条件下行驶大约需要2.5千瓦。而要让空调系统将车外的空气加热到车内的22°C并维持这个温度，则需要4.5千瓦的功率，几乎是行驶功率的两倍。这个空调热功率主要来自于汽车外表面的对流和辐射损失（约2千瓦）以及空调排出的废热（约2.5千瓦）。要维持车内外高达40度的温差，可利用的电池和动力系统的废热仅仅是0.4千瓦，这是远远不够的。此外，如果电池温度低于零度会对电池的性能和寿命造成影响。因此电池的加热在冬季也是必要的。

Im Sommer (Randbedingungen: TUmgebung = 36  °C und 25 % relative Feuchte, Luftmenge 5 kg/min, TKabine = 22 °C, A/B-Segment-Fahrzeug, vFahrzeug = 18,3 km/h) erfolgt der Hauptwärmeeintrag über solare Strahlung von bis zu 1 kW/m2 sowie über Konvektion. Der durch die Klimatisierung bedingte „Kälteverlust“ erfolgt über die Abluft aus der Kabine. Insgesamt müssen so für die Aufrechterhaltung von 22 °C Kabinentemperatur rund 3,8 kW an Kältel eistung und unter Berücksichtigung der Leistungszahl des Klimakreislaufs insgesamt etwa 2 kW von der Batterie aufgebracht werden. Diese Leistung entspricht der gleichen Größenordnung, wie sie für den Vortrieb erforderlich ist. Bei der Abkühlung nach dem Start müssen bei nicht vorkonditioniertem Fahrzeug, je nach Außentemperatur, sehr hohe Temperaturdifferenzen ausgeglichen werden, da der Fahrzeuginnenraum durch die solare Einstrahlung stark überhitzt ist.  

在夏季条件下（环境温度36°C, 25%的相对湿度，车内温度22°C，车辆速度18.3km/h，空调空气流量5kg/min），将近1千瓦/平方米的太阳辐射和对流是车辆的主要热量输入源。加上空调系统排出的废冷空气造成的“冷损失”， 为了达到并维持车内22°C的恒温需要2.8千瓦的冷却功率。经过性能系数的折算，电池需要提供大约2千瓦的功率。这和汽车行驶功率在同一个量级上的。

Für die Aufrechterhaltung der Komforttemperatur müssen dagegen nur bis zu 15 K (plus die immer noch vorhandene solare Strahlung) überbrückt werden. Im Gegensatz zur Situation im Winter muss die Batterie in der heißen Jahreszeit je nach Temperatursituation gekühlt werden. Die obere Grenztemperatur beträgt + 40 °C. Bei Temperaturen darüber wird die Batterie während des Betriebs irreversibel geschädigt; sie muss also im Sommer aktiv gekühlt werden.

由于强烈的太阳辐射，空调系统需要维持车内比车外低15度的温差。而电池在夏季需要被主动冷却以维持在40°C以下。超过了这个温度，电池在运行中将受到不可逆的损坏。

Aus diesen Szenarien ergeben sich die vielfältigen Anforderungen an das Thermomanagement. Sowohl im Winter als auch im Sommer soll der Kabinenkomfort und damit auch die Sicherheit (Scheibenbeschlag und -vereisung im Winter, Konzentrations fähigkeit im Sommer) gewährleistet werden. Ferner muss die Batter ie stets in einem relativ engen Temperaturfenster betrieben werden, um die Sicherheit des Fahrzeugs und eine möglichst lange Lebensdauer der Batterie zu gewährleisten. Da bei reinen E-Fahrzeugen die Energie für Heiz- und Kühlsysteme in direkter Konkurrenz zur elektrischen Reichweite steht, ist die Reduzierung beziehungsweise die Effizienzsteigerung der benötigten Heiz- und Kühlungsleistung extrem wichtig.

不同的使用场景对热管理提出了不同的要求：座舱和电池温度在四季都需要保持在一个合适的范围内，对此所消耗的能量又会造成车辆的续航里程的减少。所以提高热系统的效率显得越发重要。

### Elektrifizierungsgrad  und Klimatisierung des Innenraums 
电气化程度对车内空调的影响

Im rein verbrennungsmotorisch angetriebenen Fahrzeug fällt die Reichweitenreduzierung durch den Energiebedarf der Klimatisierung kaum ins Gewicht. Im Heizungsfall steht aufgrund des Wirkungsgrads des Verbrennungsmotors von lediglich 30 bis 40 % genügend Abwärme zur Verfügung, die über den Kühlmittelkreislauf an den Heizkörper der Innenraumheizung beziehungsweise an die Batterie abgegeben werden kann. Mit steigender Hybridisierung entlang des sogenannten  Elektrifizierungspfads (auch Elektrifizierungsgrad genannt, siehe Kapitel 1.1) nehmen sowohl die im Fahrzeug verfügbare Kraftstoffenergie zum Kühlen als auch die aus dem Verbrennungsmotor anfallende Wärme im Kühlmittel ab, Bild 4-15. 

在纯燃油车中，空调的使用不会显著地减少续航里程。这是因为内燃机只有30%至40%的效率，所以当空调在加热时，可以从内燃机获得足够的废热。这些废热通过冷却回路被输送到车内的暖气加热器或者用来加热电池。随着混合动力中电气化比重的提高，能够利用的内燃机的废热变得越来越少。


Schon bei Fahrzeugen mit Start-StoppAutomatik (auch Mikrohybrid bezeichnet) zeigen sich Auswirkungen auf die Klimatisierung des Innenraums. Denn durch die häufigen Abschaltzeiten des Motors sinkt zum einen der Wärmeeintrag ins Kühlmittel, zum anderen steht in diesen Phasen der Klimakompressor still, sodass der Innenraum nicht mehr gekühlt wird. 

在微混车辆中，启停系统已经可以对空调造成影响。 由于频繁的关闭发动机， 进入冷却回路的内燃机废热会显著地减少。此外由于空调压缩机也会停止工作， 所以车内冷却会受到影响。

Ab den Mildhybrid-Konzepten wird der verbrennungsmotorische Antrieb zumindest teilweise, zum Beispiel beim Beschleunigen, durch eine E-Maschine mit entsprechend dimensionierter Lithium-Ionen-Batterie unterstützt. Vor allem im Winter kann das zunehmende Downsizing des Verbrennungsmotors zu einer geringeren Abwärme und damit zu einer eingeschränkten Heizfunktion führen. Dieses Heizleistungsdefizit kann jedoch durch den vorhandenen fossilen Kraftstoff an Bord entweder direkt oder über einen elektrischen Zuheizer (PTC) ausgeglichen werden. Zur Kabinenkühlung wird der herkömmliche mechanisch angetriebene Kompressor des Klimakreislaufs in der Regel durch einen Elektrokompressor ersetzt. Damit kann Kraftstoff eingespart werden, da in den Abschaltzeiten des Motors die notwendige Kompressorantriebsenergie aus der rekuperierten Energie, gespeichert in der Lithium-Ionen-Batterie, stammt. Zudem übernimmt der Klimakreislauf nun zusätzlich die Kühlungsaufgabe für die temperaturempfindliche Lithium-IonenBatterie.

在轻度混合动力车辆中，行驶所需的动力部分地由电机提供。轻量化的内燃机产生的废热不足以完全提供车内加热的所需的能量，还需要通过直接燃油或者电加热（PTC）的方式来弥补不足。在冷却的场景中， 空调压缩机也由传统的由发动机机械拖动变为电机驱动，这样可以用储存在锂电池中的回收能量来驱动空调， 以达到省油的目的。 空调回路也可以用来冷却锂电池。

Vollhybridisierte Fahrzeuge mit und ohne Plug-in-Funktionalität, die durch eine entsprechend dimensionierte Batterie bis zu etwa 50 km rein elektrisch fahren können, verfügen ebenfalls über einen elektrisch angetriebenen Kompressor für den Klimakreislauf. Somit steht dieser sowohl im rein elektrischen als auch im verbrennungsmotorischen Antriebmodus zur Verfügung. Das Fahrzeug sollte jedoch prinzipiell so oft wie möglich elektrisch bewegt werden. 

强混和插电混合动力车辆的纯电里程可以达到50公里，为了能让空调无论在纯电或燃油驱动状态下都能工作， 空调压缩机的也是用电机驱动的。

Aufgrund der hohen Wirkungsgrade der elektrischen Antriebskomponenten E-Motor, Leistungselektronik und Batterie von jeweils bis zu 95 % fällt hierbei so gut wie keine verwertbare Abwärme zum Heizen an. Die Heizenergie muss dann der Batterie entnommen und über einen Hochvoltheizer in Wärme für den Innenraum umgesetzt werden. Dies geht allerdings zu Lasten der elektrischen Reichweite. Im Heizfall könnte jedoch situationsbedingt auf die Abwärme des Verbrennungsmotors zurückgegriffen werden. Verfügt der Vollhybrid über eine Plug-in-Funktionalität, kann das Fahrzeug im Winter und im Sommer über die Steckdose vorkonditioniert, das heißt vor Abfahrt vorgeheizt beziehungsweise vorgekühlt werden. Damit steht bei Beginn der Fahrt von der im Fahrzeug gespeicherten Energie mehr für den Fahrantrieb zur Verfügung. 

由于电机，功率电子和电池分别能够达到95%的效率，它们几乎不提供可以利用的废热。空调所需的能量必须从电池获得。 纯电的续航里程会因此减少。如果混动车辆能够用外部插座充电，在充电时调整车内的温度到适宜的水平。这样电池就能有更多的能量用于纯电行驶。

Bei einem Elektrofahrzeug mit Range Extender steht zwar prinzipiell die Abwärme des Verbrennungsmotors zur Verfügung. Da dieser jedoch in der Regel leistungsmäßig nur für die Batterieladung und nicht für den Fahrbetrieb ausgelegt ist, liefert er nur begrenzt Abwärme. Auch hier muss auf Kosten der elektrischen Reichweite elektrisch zugeheizt werden. Eine Vorkonditio n ierung über die Steckdose ist natürlich weiterhin möglich.

具有增程器的电动车原则上可以利用内燃机的废热， 但是作为增程器的内燃机往往只针对为电池充电的目的设计，它只能提供有限的热量。所以这种混动方式也必须使用电池的能量进行加热，这也将导致续航里程的减少。


In einem reinen Elektrofahrzeug steht allerdings so gut wie keine Abwärme mehr aus dem Antriebsstrang für die Innenraumheizung zur Verfügung. Die Energie zum Heizen und für den Klimakreislauf muss vollständig von der Lithium-IonenBatterie bereitgestellt werden. Eine Vorkonditionierung ist zwar möglich, dennoch verkürzen Kühlung und Heizung direkt die Reichweite des Fahrzeugs.

在纯电车辆中， 几乎就没有像内燃机那样的废热源用来加热座舱。 空调和加热所需的能量完全来自锂电池。这必然会导致续航里程的减少。


### Effizienzsteigerung der Klimatisierung   
Generell werden passive und aktive Maßnahmen zur Effizienzsteigerung der Klimatisierung unterschieden. Passive Maßnahmen sind zum Beispiel die Verminderung des solaren Wärmeeintrags im Sommer über die Scheibenflächen und die Karosserie durch eine verbesserte Kabinenisolierung und einen Strahlungsschutz der Scheiben. Eine Verringerung der thermischen Massen im Innenraum, zum Beispiel bei Sitzen, Armaturentafel oder Verkleidungen, ist ebenso hilfreich. Diese Möglichkeiten werden hier allerdings nicht weiter betrachtet.

提高空调的效率

提高空调的效率通常可以运用被动和主动的措施。 被动的措施指的是改善车厢隔热和减少太阳穿过车窗的热辐射。 同时减少车内的热质量， 如减少座椅或面板的质量， 这也会对提高空调效率有所帮助。


Moderne Klimasysteme erlauben eine thermische Vorkonditionierung und sind auf eine möglichst schnelle Abkühlung oder Aufheizung der Fahrzeugkabine ausgelegt. Durch die Vorkonditionierung über eine externe Energiequelle kann der Innenraum vor Fahrtbeginn auf die gewünschte Komforttemperatur gebracht werden. Während der Fahrt muss so die Batterie nur noch die Erhaltungsenergie zur Verfügung stellen. Dadurch wird die elektrische Reichweite erhöht. Zudem können die Komponenten des Klimasystems gegebenenfalls kleiner und damit leichter ausgelegt werden. Mittels geführter Luftströmung in der Kabine in Abhängigkeit von der Sitzbelegung kann eine effiziente, bedarfsgerechte Klimatisierung erreicht werden.

现代空调系统允许预先对车内温度尽可能快地经行调节。 通过由外部电源对车辆的预先调温， 车内在车辆启动前可以达到一个期望的舒适温度。在车辆行驶中，电池能量只被用来维持这个温度。这样就增加了续航里程。此外，通过车内空调会针对座椅的位置而精确引导的气流， 这可以使空调更有效率和更针对乘客需求。
